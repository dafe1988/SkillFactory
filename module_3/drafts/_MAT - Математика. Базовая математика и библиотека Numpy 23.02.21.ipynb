```
import matplotlib.pyplot as plt
import numpy as np
import math

plt.figure(figsize=(4, 4), dpi=150)
X = np.linspace(-2.5, 2.5, 1000)
plt.plot(X, X, label = 'y = x')
plt.plot(X, X**2, label = 'y = x^2')
plt.plot(X, X**3, label = 'y = x^3')
plt.plot(X, X**4, label = 'y = x^4')
xlims = (-2.5, 2.5)
ylims = (-5, 5)
plt.gca().set_xlim(*xlims)
plt.gca().set_ylim(*ylims)
plt.xticks(np.arange(math.ceil(xlims[0]), math.floor(xlims[1])+1, 1))
plt.yticks(np.arange(math.ceil(ylims[0]), math.floor(ylims[1])+1, 1))
plt.grid()
plt.legend()
plt.show()
```

```
plt.figure(figsize=(4, 4), dpi=150)
X = np.linspace(-2.5, 2.5, 1000)
plt.plot(X, X**2+X, label = 'y = x^2+x')
plt.plot(X, X**4-2*X**3-X, label = 'y = x^4-2x^3-x')
plt.plot(X, -X**6+X**3+3*X**2+1, label = 'y = -x^6+x^3+3*x^2+1')
xlims = (-2.5, 2.5)
ylims = (-5, 5)
plt.gca().set_xlim(*xlims)
plt.gca().set_ylim(*ylims)
plt.xticks(np.arange(math.ceil(xlims[0]), math.floor(xlims[1])+1, 1))
plt.yticks(np.arange(math.ceil(ylims[0]), math.floor(ylims[1])+1, 1))
plt.grid()
plt.legend()
plt.show()
```

```
plt.figure(figsize=(6, 6), dpi=100)
X = np.linspace(-2.5, 2.5, 1000)
plt.plot(X, X**3, label = 'y = x^3')
plt.plot(X, X**2, label = 'y = x^2')
plt.plot(X, X**1.5, label = 'y = x^1.5')
plt.plot(X, X**1, label = 'y = x^1')
plt.plot(X, X**0.5, label = 'y = x^1/2')
plt.plot(X, X**(1/3), label = 'y = x^1/3')
plt.plot(X, X**0, label = 'y = x^0')
plt.plot(X, X**-1, label = 'y = x^-1')
plt.plot(X, X**-2, label = 'y = x^-2')
xlims = (-2.5, 2.5)
ylims = (-5, 5)
plt.gca().set_xlim(*xlims)
plt.gca().set_ylim(*ylims)
plt.xticks(np.arange(math.ceil(xlims[0]), math.floor(xlims[1])+1, 1))
plt.yticks(np.arange(math.ceil(ylims[0]), math.floor(ylims[1])+1, 1))
plt.grid()
plt.legend()
plt.show()
```

```
plt.figure(figsize=(8, 5), dpi=150)
xlims = (-16, 16)
ylims = (-1.7, 1.7)
X = np.linspace(*xlims, 1000)
plt.plot(X, 1/(1+np.exp(-X)), label = 'y = sigmoid(x)')
plt.plot(X, np.arctan(X), label = 'y = arctan(x)')
plt.plot(X[np.where(X < 0)], np.sign(X[np.where(X < 0)]), label = 'y = sign(x)')
plt.plot(X[np.where(X > 0)], np.sign(X[np.where(X > 0)]), color = 'C2')
plt.gca().set_xlim(*xlims)
plt.gca().set_ylim(*ylims)
plt.xticks(np.arange(math.ceil(xlims[0]), math.floor(xlims[1])+1, 2))
plt.yticks(np.arange(math.ceil(ylims[0]), math.floor(ylims[1])+1, 1))
plt.grid()
plt.legend()
plt.show()
```
