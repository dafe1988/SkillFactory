```python
log2['hour'] = log2.time.dt.hour

def day_time(x):
    x=int(x)
    if 0<=x<=5:
        return 'night'
    elif 6<=x<=11:
        return 'morning'
    elif 12<=x<=17:
        return 'day'
    else:
        return 'evening'
    
log2.hour = log2.hour.apply(day_time)
log2
```

```python
log = pd.read_csv('log.csv', header=None)
log.columns = ['user_id', 'time', 'bet', 'win']
log = log.dropna()
log['time'] = pd.to_datetime(log['time'].apply(lambda x: x[1:]))
log['hour'] = log.time.dt.hour
log
```

```python
log = pd.read_csv('log.csv', header=None)
log.columns = ['user_id', 'time', 'bet', 'win']
log
```

```python
log.dtypes
```

```python
log["win_"] = np.where(log["win"] > 0, log['win'], -log['bet'])
log
```

```python
def fillna_win(row):
    if row.win > 0:
        return row.win
    elif row.bet > 0:
        return -row.bet
    return 0
        
new_win = log.apply(lambda row:fillna_win(row),axis=1)
log['win'] = new_win
log
```

```python
log[log['win']<0].win.value_counts().sum()
```

```python
def net_(row):
    if row.win < 0:
        return row.win
    return (row.win-row.bet)

log.bet = log.bet.fillna(0)
log['net'] = log.apply(lambda row:net_(row),axis=1)
log
```

```python
log[log.net>0].net.value_counts().sum()
```
