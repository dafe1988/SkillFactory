```python
import pandas as pd
import cufflinks as cf
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
%matplotlib inline
```

```python
init_notebook_mode(connected = True)
cf.go_offline()
```

```python
df = pd.read_csv('tips.csv')
```

```python
df.iplot()
```

```python
df[['total_bill', 'tip']].iplot(kind='spread')
```

```python
df.iplot(kind='bubble', x='total_bill', y='size', size='tip')
```

```python
import plotly.express as px

df = px.data.tips()
fig = px.violin(df, y="total_bill")
fig.show()
```

```python
fig = px.violin(df, y="total_bill", box=True, # draw box plot inside the violin
                points='all', # can be 'outliers', or False
               )
fig.show()
```

```python
fig = px.violin(df, y="tip", x="smoker", color="sex", box=True, points="all",
          hover_data=df.columns)
fig.show()
```

```python
fig = px.parallel_categories(df)

fig.show()
```

```python
fig = px.parallel_categories(df, dimensions=['sex', 'smoker', 'day'],
                color="size", color_continuous_scale=px.colors.sequential.Inferno,
                labels={'sex':'Payer sex', 'smoker':'Smokers at the table', 'day':'Day of week'})
fig.show()
```

```python
df.pivot_table(index = 'day', columns = 'sex', values = 'total_bill', aggfunc = 'mean').iplot(kind = 'bar') 
```

```python
df.groupby(by = 'day')['total_bill'].agg(['min','max']).plot() 
```
