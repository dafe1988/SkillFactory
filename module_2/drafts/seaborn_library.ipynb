```python
import numpy as np
import pandas as pd
import scipy.stats as sps

import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
def sinplot(flip=1):
    x = np.linspace(0, 14, 100)
    for i in range(1, 7):
        plt.plot(x, np.sin(x + i * .5) * (7 - i) * flip)
```

```python
sinplot()
```

```python
plt.figure(figsize=(15, 9))
for i, context in enumerate(['notebook', 'paper', 
                             'talk', 'poster']):
    sns.set(context=context)  # Устанавливаем стиль
    plt.subplot(2, 2, i+1)
    sinplot()
    plt.title(context)
```

```python
plt.figure(figsize=(15, 12))
for i, style in enumerate(['darkgrid', 'whitegrid', 
                           'dark', 'white', 'ticks']):
    sns.set(style=style)  # Устанавливаем стиль
    plt.subplot(3, 2, i+1)
    sinplot()
    plt.title(style)
```

```python
sns.set(style='white')
sinplot()
sns.despine(fig=None, ax=None, top=True, right=True, left=False, bottom=False, offset=None, trim=False)
'''убрать рамку -- fig — фигура matplotlib. Если None, то текущая;
    top, right, left, bottom — указатели границ. 
    Левую и нижнюю оставляют, т.к. вдоль них расположены метки на координатных осях.'''
```

```python
with sns.plotting_context("notebook"), sns.axes_style("ticks"):
    sinplot()
    sns.despine()
```
