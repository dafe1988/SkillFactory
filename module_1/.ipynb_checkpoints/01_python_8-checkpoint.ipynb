```python
sample = pd.read_csv("sample.csv")
sample.Age = sample.Age.apply(lambda x:x+1)
sample2 = sample
sample2
```

```python
sample = pd.read_csv("sample.csv")
sample.City = sample.City.apply(lambda s: str(s).lower())
sample2 = sample
sample2
```

```python
sample = pd.read_csv("sample.csv")

def profession_code(s):
    if s == 'Рабочий':
        return 0
    elif s == 'Менеджер':
        return 1
    else:
        return 2

sample2 = sample    
sample2.Profession = sample2.Profession.apply(profession_code)
sample2
```

```python
sample = pd.read_csv("sample.csv")

def age_category(age):
    if age<23:
        return "молодой"
    elif 23<age<35:
        return "средний"
    elif age>35:
        return "зрелый"
    
sample['Age_category'] = sample.Age.apply(age_category)
sample
```

```python
log = pd.read_csv('log.csv', header=None)
log.columns = ['user_id','time','bet','win']

def format_user(user_id: str):
    user_id = user_id.replace('Запись пользователя № - ','')
    user_id = user_id.replace('#error','')
    return user_id

log.user_id = log.user_id.apply(format_user)
log
```

```python
log = pd.read_csv('log.csv', header=None)
log.columns = ['user_id','time','bet','win']
log.time = log.time.str.strip('[')
log
```

```python
log = pd.read_csv("log.csv",header=None) 
log.columns = ['user_id','time','bet','win']

def format_t(time):
    time = str(time).strip('[')
    return time

log.time = log.time.apply(format_t)
t = log.time[0]
t
```

```python
# прочитаем файл в переменную log
log = pd.read_csv('log.csv', header=None)
# добавим названия колонок user_id, time, bet, win
log.columns = ['user_id','time','bet','win']
# удалим строки, которые содержат значения user_id с ошибками
log = log[~log.user_id.str.contains('#error',na=False)]
# оставим в поле user_id значение типа: "user_N", где N значение идентификатора
log.user_id = log.user_id.str.strip('Запись пользователя № - ')
# уберём начальную скобку из поля time
log.time = log.time.str.strip('[')

log
```

```python
log = pd.read_csv('log.csv', header=None)
log.columns = ['user_id','time','bet','win']
log = log[~log.user_id.str.contains('#error',na=False)]
log.user_id = log.user_id.str.strip('Запись пользователя № - ')
log.time = log.time.str.strip('[')
log
```
